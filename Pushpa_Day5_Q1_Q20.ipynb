# 12.12.Create a visualization on the count of patients based on primary diagnosis.¶

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel('HospitalDatabase.xlsx', "ReAdmissionRegistry")
patient_counts = df.groupby('PrimaryDiagnosis')['PatientID'].count()
plt.pie(patient_counts)
my_circle = plt.Circle((0,0), 0.5, color='white')
plt.gca().add_artist(my_circle)
plt.title('Patient Count in ReAdmissionRegistry')
plt.show()
```

# 9.Create a subplot on x = np.arange(0, 10, 0.1) , y = np.sin(np.pi * x) + x with 2 rows & 2 columns with object oriented approach.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#asigning values to x & y
x = np.arange(0,10,0.1)
y = np.sin(np.pi*x)+x

#Simple plot
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title('simple plot')

#plot x with 2 rows & 2 colomns
rows, cols = 2,2
fig, ax = plt.subplots(rows, cols, sharex='col',sharey='row')
for row in range(rows):
    for col in range(cols):
        ax[row, col].text(0.5, 0.5, str((row, col)))
```

# 12.Create a visualization on the count of patients based on primary diagnosis.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel('HospitalDatabase.xlsx', "ReAdmissionRegistry")
grouped_df = df.groupby('PrimaryDiagnosis')['PatientID'].count().reset_index()
plt.bar(grouped_df['PrimaryDiagnosis'], grouped_df['PatientID'])
plt.xticks(rotation=90)
plt.xlabel('Primary Diagnosis')
plt.ylabel('Count of Patients')
plt.title('Count of Patients by Primary Diagnosis')
plt.show()
```

# 13.Create a bar chart on service & expected length of stay.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel('HospitalDatabase.xlsx', "ReAdmissionRegistry")
grouped_df = df.groupby('Service')['ExpectedLOS'].mean().reset_index()
plt.bar(grouped_df['Service'], grouped_df['ExpectedLOS'], width = .40)
plt.xticks(rotation=90)
plt.xlabel('Service')
plt.ylabel('Expected Length of Stay')
plt.title('Expected Length of Stay by Service')
plt.show()
```
